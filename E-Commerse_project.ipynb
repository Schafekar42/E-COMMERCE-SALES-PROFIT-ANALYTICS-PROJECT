```python
# import liabraries

import pandas  as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Step -1 Loading Data set

Sales_df = pd.read_excel("C:/Users/charp/OneDrive/Desktop/Data Analytics projects/Amazon E-commerse project/ECOMM DATA-2..xlsx")


Sales_df.head()
```

```python
# Data infromation - series/Dataframe
print(Sales_df.info())
```

```python
# Satatistical Summary
print(Sales_df.describe)
```

```python
# Column Names 
print(Sales_df.columns)
```

## Data Cleaning 

```python
print(Sales_df.isnull().sum())
```

```python
# rows with null values
print(Sales_df[Sales_df.isnull().any(axis = 1)])
```

```python
## droping null rows 
Sales_df = Sales_df.dropna()
```

```python
print(Sales_df.isnull().sum())
```

```python
## Check duplicate rows
print(Sales_df.duplicated().sum())
```

```python
## Checking Data type 
print(Sales_df['Order Date'].dtype)
```

```python
# Creating Year column
Sales_df['Year'] = Sales_df['Order Date'].dt.year
```

```python
print(Sales_df['Year'])
```

```python
# creating Month Column
Sales_df['Month'] = Sales_df['Order Date'].dt.month_name()
```

```python
print(Sales_df['Month'])
```

## KPI Analysis

### Total Sales 

```python
print(Sales_df.info())
```

```python
Total_Sales = Sales_df['Sales'].sum()
print("Total sales:",Total_Sales)
```

### Total Profit

```python
Total_Profit = Sales_df['Profit'].sum()
print("Total Profit:", Total_Profit)
```

### Total Orders 

```python
Total_orders = Sales_df['Order ID'].nunique()
print("Total orders:", Total_orders)
```

### Total Quantity Sold

```python
Total_Qty = Sales_df['Quantity'].sum()
print("Total Quantity Sold:", Total_Qty)
```

### Profit Margin % 

```python
Profit_Margin = (Total_Profit / Total_Sales)*100
print("profit Margin:", Profit_Margin)
```

## Visualization 

#### Monthly Sales Trend

```python
Monthly_sales = Sales_df.groupby("Month")["Sales"].sum()
print(Monthly_sales)
```

## Line Chart 

```python
Monthly_sales.plot(kind ='line',marker ='*', figsize = (10,7))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

plt.show()
```

#### Sales By Category

```python
Sales_by_Category = Sales_df.groupby('Category')['Sales'].sum().sort_values(ascending = False)
```

```python
print(Sales_by_Category)
```

```python
Sales_by_Category.plot(kind = 'barh', figsize = (8,6))
plt.title("Sales by Category")
plt.xlabel("Sales")
plt.ylabel("Category")

plt.show()
```

#### TOP 10 Products

```python
Top_products = Sales_df.groupby("Product Name")["Sales"].sum().sort_values(ascending = False).head(10)

print(Top_products)
```

#### Horizontal Bar chart

```python
Top_products.plot(kind='barh',figsize=(10,10))
plt.gca().invert_yaxis() ## to show the top highest value at top
plt.title("TOP 10 Product by Sales")
plt.xlabel("Sales")
plt.ylabel("Product Name")

plt.show()
```

### Region Wise Sales 

```python
## for creating map 
import plotly.express as px 
```

```python

country_sales = Sales_df.groupby('Country')['Sales'].sum().reset_index()

fig = px.choropleth(
    country_sales,
    locations='Country',
    locationmode='country names',
    color='Sales',
    hover_name='Country',
    color_continuous_scale='Blues',
    title='Country Wise Sales Map'
)

# Increase map size

fig.update_layout(
    width=1200,
    height=700
)

fig.show()
```

# Advance Analytics

#### Profit Vs Discount

```python
# Converting Discount in %
Sales_df['Discount %']= (Sales_df["Discount"]*100).round(0)

Profit_Vs_Discount = Sales_df.groupby("Discount %")["Profit"].sum()

Profit_Vs_Discount_M = Profit_Vs_Discount/ 1000000

print(Profit_Vs_Discount_M)

Profit_Vs_Discount_M.plot(kind ='bar',figsize=(10,8))
plt.title("Profit Vs Discount")
plt.xlabel("Discount %")
plt.ylabel("Profit (Millions)")
plt.xticks(rotation=0)

plt.show()
```

### Shipping Cost Vs Profit

```python
plt.figure(figsize=(8,6))
plt.scatter(Sales_df["Shipping Cost"], Sales_df["Profit"])
plt.title("Profit Vs Shipping Cost")
plt.xlabel(" Shipping Cost")
plt.ylabel("Profit")

plt.show()
```

### Customer Segment Analysis

```python
Segment_Sales = Sales_df.groupby("Segment")["Sales"].sum()
print(Segment_Sales)
```

#### Using Donut Chart

```python
Segment_Sales.plot(kind ='pie', autopct='%1.1f%%',labeldistance= 0.45, figsize=(7,7))
plt.legend(title = "Segment_Sales",loc = "center right",bbox_to_anchor=(1.3, 0.5))
plt.title("Customer Segment Analysis")

plt.show()

```

### Profit by Category

```python
profit_by_Category = Sales_df.groupby("Category")["Profit"].sum()
print(profit_by_Category)
```

```python
profit_by_Category.plot(kind = 'bar', figsize=(8,6))
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation = 0)

plt.show()
```
